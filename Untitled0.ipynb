```python
path=("/content/drive/MyDrive/Classroom/Tutuorial1.csv")
df=pd.read_csv(path)
df0=df[df['Defaulter_Status']==0]
df1=df[df['Defaulter_Status']==1]
df0.describe()
```

```python
df1.describe()
```

```python
meandf0age=df0['Age'].mean()
meandf0Inc=df0['Income'].mean()
meandf0CA=df0['Credit_Amount\n'].mean()
meandf0E=df0['Education'].mean()

print(meandf0age)
print(meandf0Inc)
print(meandf0CA)
print(meandf0E)
```

```python
from scipy import stats
#it will trim your data at topmost and bottommost by 10%
tmeandf0age=stats.trim_mean(df0['Age'],0.1)
tmeandf0Inc=stats.trim_mean(df0['Income'],0.25)
tmeandf0CA=stats.trim_mean(df0['Credit_Amount\n'],0.1)
tmeandf0E=stats.trim_mean(df1['Education'],0.1)
print(tmeandf0age)
print(tmeandf0Inc)
print(tmeandf0CA)
print(tmeandf0E)
```

```python
import seaborn as sns
# You build it on top of the matplotlib data visualization library and can perform exploratory analysis
plt.subplot(2,2,1)
sns.boxplot(x='Defaulter_Status',y='Age',data=df)
plt.title('Age')
plt.subplot(2,2,2)
sns.boxplot(x='Defaulter_Status',y='Income',data=df)
plt.title('Income')
plt.subplot(2,2,3)
sns.boxplot(x='Defaulter_Status',y='Education',data=df)
plt.title('Edu')
plt.subplot(2,2,4)
sns.boxplot(x='Defaulter_Status',y='Credit_Amount\n',data=df)
plt.title('Credit_Amount')

plt.tight_layout()
plt.show()
```

```python
plt.hist([df0['Age'],df1['Age']],bins=10,label=['Defaulter_Status 0','Defaulter_Status 1'])
plt.xlabel('Age')
plt.ylabel('counts')
plt.legend()
plt.title("age by defaulter status")

```

```python
# Education Level by Defaulter Status
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 4))
education_counts = pd.crosstab(df['Defaulter_Status'], df['Education'])
education_counts.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_xlabel('Defaulter_Status')
axes[0].set_ylabel('Counts')
axes[0].set_title('Education Level by Defaulter Status')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['0', '1'], rotation=0)

employed_counts = pd.crosstab(df['Defaulter_Status'], df['Employed'])
employed_counts.plot(kind='bar', stacked=False, ax=axes[1])
axes[1].set_xlabel('Defaulter_Status')
axes[1].set_ylabel('Counts')
axes[1].set_title('Employed by Defaulter Status')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['0', '1'], rotation=0)
```
